```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(pd.read_csv('D3.csv'))
df.head()
```

```python
def cost(x, y, theta):
    h = x.dot(theta)
    E = np.subtract(h, y)
    sqrE = np.square(E)
    J = 1/(2*m)*np.sum(sqrE)
    
    return J

def gradient_descent(x, y, theta, alpha, iterations):
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        h = x.dot(theta)
        e = np.subtract(h, y)
        sum_delta = (alpha/m)*x.transpose().dot(e);
        theta = theta - sum_delta;
        cost_history[i] = cost(x, y, theta)
    
    return theta, cost_history
```

```python
def training(x, theta, iterations, alpha):
    x = x.reshape(m, 1)
    x = np.hstack((np.ones((m, 1)), x))

    theta, cost_history = gradient_descent(x, y, theta, alpha, iterations)

    plt.scatter(x[:,1], y, color='red', marker= '+', label= 'Training Data')
    plt.plot(x[:,1], x.dot(theta), color='green', label='Linear Regression')
    plt.grid()
    plt.title('Linear Regression Fit')
    plt.legend()
    plt.show()

    plt.plot(range(1, iterations+1),cost_history, color='blue')
    plt.grid()
    plt.xlabel('Number of iterations')
    plt.ylabel('Cost')
    plt.title('gradient descent')
    plt.show()
```

```python
dataset = df.values[:,:]
y = df.values[:, 3]
m = len(y)
theta = [0, 0]
iterations = 2000;
alpha = 0.01;
```

```python
# Training for X1
training(df.values[:, 0], theta, iterations, alpha)
```

```python
# Training for X2
training(df.values[:, 1], theta, iterations, alpha)
```

```python
# Training for X3
training(df.values[:, 2], theta, iterations, alpha)
```
